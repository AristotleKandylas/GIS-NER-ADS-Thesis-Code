# Unified core concept tags creation

```python
core_df= pd.read_csv("/content/drive/MyDrive/UTRECHT UNIVERISTY/THESIS/DATA/core_concept_dictionary.csv", encoding='unicode_escape')
core_df.head()
```

```python
len(core_df.CCT.unique())
```

```python
core_df.CCT.unique()
```

```python
coredf = core_df.copy()
coredf.head()
```

```python
#create core concept type abbreviation
ncct = []
for i in coredf.CCT[0:665]:
  if i =="object" or i=='Object':
    ncct.append("OBJ")
  elif i== "event" or i=="Event":
    ncct.append("EVE")
  elif i=="field":
    ncct.append("FLD")
  elif i=="network":
    ncct.append("NET")
  elif i=="object quality":
    ncct.append("OBJQ")
  elif i=="ammount":
    ncct.append("A")
  elif i=="proportion":
    ncct.append("PROP")
  elif i=="location":
    ncct.append("LOC")
  elif i=="conAmount":
    ncct.append("CNA")
  elif i=="covAmount":
    ncct.append("CVA")
  elif i=="Object conAmount":
    ncct.append("OBJCNA")
  elif i=="event quality":
    ncct.append("EVEQ")
  elif i=="CovPro":
    ncct.append("CVPR")
  elif i=="Event conAmount":
    ncct.append("EVECNA")
  elif i=="objconobjconpro":
    ncct.append("OCOCNP")
  elif i=="objconobjcovpro":
    ncct.append("OCOCVP")
  elif i=="eveconobjconpro":
    ncct.append("EVECOCNP")
  elif i=="object conAmount":
    ncct.append("OBJCNA")
  else:
    ncct.append("EVECOCVP")

coredf["N_CCT"] = ncct
coredf.head()
```

```python
len(core_df.MSRLV.unique())
```

```python
core_df.MSRLV.unique()
```

```python
#create measurement level abbreviation
msr = []
for i in coredf.MSRLV[0:665]:
  if i =="interval":
    msr.append("I")
  elif i== "nominal" :
    msr.append("N")
  elif i=="count":
    msr.append("C")
  elif i=="ratio":
    msr.append("R")
  elif i=="boolean":
    msr.append("B")
  elif i=="loc":
    msr.append("L")
  elif i=="ordinal":
    msr.append("O")
  elif i=="ERA":
    msr.append("ER")
  elif i=="IRA" or i=="ira":
    msr.append("IR")
  else:
    msr.append(" ")

coredf["N_MSRLV"] = msr
coredf.head()    
```

```python
# create unified tags from CCT and MSRLV
coredf["U_Tag"] = coredf['N_CCT']+ coredf['N_MSRLV'] 
coredf.head()
```

```python
coredf.head(30)  
```

```python
coredf2 = coredf.copy()
coredf2
```

```python
# keep only noun-noun phrases  and U_tags
coredf2 = coredf2.drop(coredf2.iloc[:, 1:5], axis=1)
```

```python
# preserve noun phrases as unified
coredf2.head()
```

```python
# export new core concept dictionary as csv file
coredf2.to_csv(r'core_concept_u_tags.csv', index=False)
```

```python
# export clean questions as txt file
coredf2.to_csv(r'core_conceot_u_tags.txt', header=None, index=None, doublequote = False, sep='\t', mode='w')
```
